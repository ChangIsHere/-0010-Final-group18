```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
enrollment_data = pd.read_csv('[Dataset3] Enrollment by School and Neighborhood.csv')
display(enrollment_data.head())
```

```python
total_enrollment_by_school = enrollment_data.groupby('school')['enrolled_students'].sum().sort_values(ascending=False)
plt.figure(figsize=(14, 8))
total_enrollment_by_school.plot(kind='bar', color='skyblue')
plt.title('Total Enrollment by School in Pittsburgh')
plt.xlabel('School')
plt.ylabel('Total Enrolled Students')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
total_enrollment_by_neighborhood = enrollment_data.groupby('neighborhood')['enrolled_students'].sum().sort_values(ascending=False)
plt.figure(figsize=(14, 8))
total_enrollment_by_neighborhood.plot(kind='bar', color='coral')
plt.title('Total Enrollment by Neighborhood in Pittsburgh')
plt.xlabel('Neighborhood')
plt.ylabel('Total Enrolled Students')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
crime_data = pd.DataFrame({
    'neighborhood': ['Brighton Heights', 'Elliot', 'Morningside', 'Crafton Heights - Westwood - Oakwood', 'Harpen Hilltop'],
    'crime_score': [0.4, 0.5, 0.3, 0.2, 0.1]  
})
pa_data = pd.DataFrame({
    'neighborhood': ['Brighton Heights', 'Elliot', 'Morningside', 'Crafton Heights - Westwood - Oakwood', 'Harpen Hilltop'],
    'number_general_assistance': [69.0, 52.0, 23.0, 26.0, 10.0]
})
enrollment_data = pd.DataFrame({
    'neighborhood': ['Brighton Heights', 'Elliot', 'Morningside', 'Crafton Heights - Westwood - Oakwood', 'Harpen Hilltop'],
    'enrolled_students': [54.363636, None, 29.833333, None, None]
})
weights = {
    'crime_score': 0.5,
    'pa_score': 0.3,
    'enrollment_score': 0.2
}
crime_data['crime_score'] = 1 - crime_data['crime_score']
final_scores = crime_data.merge(pa_data, on='neighborhood', how='left')
final_scores = final_scores.merge(enrollment_data, on='neighborhood', how='left')
final_scores['enrolled_students'] = final_scores['enrolled_students'].fillna(0)
scaler = MinMaxScaler()

final_scores['pa_score'] = scaler.fit_transform(final_scores[['number_general_assistance']])
final_scores['enrollment_score'] = scaler.fit_transform(final_scores[['enrolled_students']])
final_scores['total_score'] = (final_scores['crime_score'] * weights['crime_score'] +
                               final_scores['pa_score'] * weights['pa_score'] +
                               final_scores['enrollment_score'] * weights['enrollment_score'])

final_scores = final_scores.sort_values('total_score', ascending=False).reset_index(drop=True)
top_5_final_scores = final_scores.head(5)
top_5_final_scores
```
