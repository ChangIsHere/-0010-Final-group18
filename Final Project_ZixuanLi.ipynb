## Introduction:
Safety is often considered one of the most important indicators for assessing the quality of a residential area. When deciding where to live, individuals and families consider safety first, as a safe environment is the basis for the physical and mental health of residents and the security of their property. A low crime rate in a community indicates that residents can live and work without fear of personal safety or property damage, thus improving their quality of life and residential satisfaction.

In addition, the safety of a community has a direct impact on its attractiveness and real estate values. Safe neighborhoods are more likely to attract business investment and quality educational resources, promoting economic development and social stability. This positive feedback loop further enhances the overall competitiveness of the community, making it a desirable location for more people to choose to live.

Therefore, in this project, we consider safety as a core indicator of the "best neighborhoods". This indicator is measured by three sub-indicators: educational attainment, crime rate, and social safety. The three datasets from the WPRDC used for the evaluation are: school enrollment, crime index, and social assistance rates.

# Metric 1: Arrests for Major Crimes

## Introduction
In this analysis, we explore the landscape of major crimes across the neighborhoods of Pittsburgh. Our aim is to uncover patterns and insights that could aid in crime prevention and safety improvements across the city.

## Dataset Overview
The dataset includes information on the rate and number of arrests for major crimes, such as murder, rape, robbery, assault, and burglary. By evaluating these figures, we hope to identify neighborhoods with higher safety challenges and recognize those with effective crime prevention measures.


```python
# Import necessary libraries for analysis and visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Arrests for Major Crimes.csv')

# Display the first few rows to verify that the dataset is loaded correctly
display(data.head())
```

First five rows of dataset.



## Overall Crime Rate by Neighborhood


```python
# Visualize the overall crime rate by neighborhood
plt.figure(figsize=(14, 8))
sns.barplot(x='neighborhood', y='overall_crime_rate', data=data)
plt.xticks(rotation=90)
plt.title('Overall Crime Rate by Neighborhood in Pittsburgh')
plt.xlabel('Neighborhood')
plt.ylabel('Overall Crime Rate')
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping
plt.show()
```

## Detailed Crime Analysis
We delve deeper into specific crimes, examining patterns for each major crime category. This will help in identifying specific areas of concern for each neighborhood.



```python
# Example: Analyze the crime rate for murders
plt.figure(figsize=(14, 8))
sns.barplot(x='neighborhood', y='crime_rate_murder', data=data)
plt.xticks(rotation=90)
plt.title('Murder Crime Rate by Neighborhood in Pittsburgh')
plt.xlabel('Neighborhood')
plt.ylabel('Murder Crime Rate')
plt.tight_layout()
plt.show()

# Similar plots and analyses can be conducted for rape, robbery, assault, and burglary
```

## Statistical Summary

```python
import pandas as pd

# Replace 'your_file_path.csv' with the actual path to your CSV file.
file_path = 'Arrests for Major Crimes.csv'
data = pd.read_csv(file_path)

# Now you can display the statistical summary of numerical columns
display(data.describe())
```

## Correlation Analysis
Understanding the relationship between different types of crimes can provide insights into crime patterns. For instance, neighborhoods with high rates of one type of crime might also experience high rates of another.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Compute and visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix of Crime Data')
plt.show()

```

## Crime Rate Distribution
Visualizing the distribution of crime rates helps to understand the spread and skewness of crime data across neighborhoods.

### Distribution of Murder Crime Rates


```python
# Histogram for the distribution of murder crime rates
plt.figure(figsize=(10, 6))
sns.histplot(data['crime_rate_murder'], bins=20, kde=True, color='purple')
plt.title('Distribution of Murder Crime Rates in Pittsburgh Neighborhoods')
plt.xlabel('Murder Crime Rate')
plt.ylabel('Number of Neighborhoods')
plt.show()
```

## Crime Rates in Relation to Population
We will examine whether there's a correlation between the neighborhood population and the crime rates.

### Scatter Plot: Population vs. Overall Crime Rate


```python
# Scatter plot to show the relationship between population and overall crime rate
plt.figure(figsize=(12, 7))
sns.scatterplot(x='population', y='overall_crime_rate', data=data)
plt.title('Population vs. Overall Crime Rate in Pittsburgh Neighborhoods')
plt.xlabel('Population')
plt.ylabel('Overall Crime Rate')
plt.show()
```

## Most and Least Safe Neighborhoods
To further focus our analysis, we identify the neighborhoods with the highest and lowest crime rates.

### Top 5 Neighborhoods with Highest Overall Crime Rates


```python
# Display top 5 neighborhoods with highest overall crime rates
top_5_high_crime = data.sort_values(by='overall_crime_rate', ascending=False).head(5)
display(top_5_high_crime[['neighborhood', 'overall_crime_rate']])
```

### Top 5 Neighborhoods with Lowest Overall Crime Rates


```python
# Display top 5 neighborhoods with lowest overall crime rates
top_5_low_crime = data.sort_values(by='overall_crime_rate', ascending=True).head(5)
display(top_5_low_crime[['neighborhood', 'overall_crime_rate']])
```
